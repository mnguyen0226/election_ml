## I. Data Processing

```python
# import general libraries
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
import collections
import sortedcontainers
```

```python
# import dataset
df = pd.read_csv("../data/edit_summary_stats.csv")
df.head()
```

```python
print(f"The dimension of the dataset is: {df.shape}")
```

```python
# check data types of each columns
df.dtypes
```

```python
# select the rows in the dataset that have the voting labels (years 2008, 2012, 2016, 2020)
selected_df = df.loc[df['year'].isin([2008, 2012, 2016, 2020])]

# reset index
selected_df = selected_df.reset_index(drop=True)

# view the dataset
selected_df.head()
```

```python
print(f"The dimension of the dataset is: {selected_df.shape}")
```

```python
# Model-dev 1: Check for data error or anomalies (no need - already did in data preprocessing in another notebook)
pass 

# Model-dev 2: Examine and replace missing values 
print(selected_df.isnull().values.any())

# Model-dev 3: Add computation field (no need - already combined 2 datasets to get this big dataset)
pass 
```

**Note:** As we consider each row as a data point to our ML model, we will drop the categorical state names and abbreviation. We will also drop the all the number of votes labels except for "winner" columns (0 for democrats and 1 for republican). We don't concern other parties as they do not have a history of winning anyway.

**Drop**:
- county_fips: as there is no correlation between fips and vote numbers.
- state_po: as there is no correlation between state appreviation and vote numbers.
- county_name: as there is no correlation between county names and vote numbers.
- democrat: as we do classification task.
- green: as we do classification task.
- liberitarian: as we do classification task.
- other: as we do classification task.
- republican: as we do classification task.

```python
# Model-dev 4: Feature selection
more_selected_df = selected_df.drop(columns=['county_fips', 'state_po', 'county_name', 'democrat',
                                             'green', 'liberitarian', 'other', 'republican'])

more_selected_df.head()
```

```python
# fill the NaN with both forward and backward fills
more_selected_df = more_selected_df.fillna(method='ffill')
more_selected_df = more_selected_df.fillna(method='bfill')
```

```python
# check if there is any NaN values
print(more_selected_df.isnull().values.any())
```

**Note:** Here, we have the complete the process of the dataset and we jave the dataset containing only counties that vote in election years in 2008, 2012, 2016 and 2022

## Data Exploration
**Note:** Let's check the count of the labels in the dataset

```python
# get the last column
winner_arr = more_selected_df['winner']
```

```python
# go thru the last column and count 1 or 0, then plot the bar graph
def count_label(arr):
    label_count = sortedcontainers.SortedDict()

    for label in arr:
        if label not in label_count:
            label_count[label] = 0
        else:
            label_count[label] += 1
    return label_count
```

```python
# get the votes count in both parties for 4 election years
label_count_all = count_label(winner_arr)
```

**Note:** Here, we see that the total number of votes for both parties are equal. Thus, this is a balanced dataset. Good for training. 0 is democrates votes and 1 is republican votes.

```python
# plot the dictionary to bar graph
votes = [label_count_all[0], label_count_all[1]]
names = ['# of Democrat Votes', '# of Republican Votes']

plt.bar(names, votes, color=['blue', 'red'])
plt.title('Total Number of Votes of Two Parties in 2008, 2012, 2016, and 2020')
plt.xlabel('Parties')
plt.ylabel('Number of Votes at Each County')
plt.show()
```

**Note:** We then count the vote labels for all 4 years

```python
# get the votes in 2008
votes_2008 = more_selected_df.loc[more_selected_df['year'] == 2008]['winner']

# get the votes count in both parties in 2008
label_count_2008 = count_label(votes_2008)
label_count_2008
```

```python
# get the votes in 2012
votes_2012 = more_selected_df.loc[more_selected_df['year'] == 2012]['winner']

# get the votes count in both parties in 2012
label_count_2012 = count_label(votes_2012)
label_count_2012
```

```python
# get the votes in 2016
votes_2016 = more_selected_df.loc[more_selected_df['year'] == 2016]['winner']

# get the votes count in both parties in 2016
label_count_2016 = count_label(votes_2016)
label_count_2016
```

```python
# get the votes in 2020
votes_2020 = more_selected_df.loc[more_selected_df['year'] == 2020]['winner']

# get the votes count in both parties in 2020
label_count_2020 = count_label(votes_2020)
label_count_2020
```

```python
votes = [label_count_2008[0], label_count_2008[1], 
         label_count_2012[0], label_count_2012[1], 
         label_count_2016[0], label_count_2016[1],
         label_count_2020[0], label_count_2020[1]]
names = ['# of Democrat Vote 2008', '# of Republican Vote 2008', 
         '# of Democrat Vote 2012', '# of Republican Vote 2012',
         '# of Democrat Vote 2016', '# of Republican Vote 2016',
         '# of Democrat Vote 2020', '# of Republican Vote 2020',]

plt.figure(figsize=(20, 5))
plt.bar(names, votes, color=['blue', 'red', 'blue', 'red', 'blue', 'red', 'blue', 'red'])
plt.title('Total Number of Votes of Two Parties in 2008, 2012, 2016, and 2020')
plt.xlabel('Parties')
plt.ylabel('Number of Votes at Each County')
plt.show()
```
