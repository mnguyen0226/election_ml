## I. Process Election Results Dataset
- Note: I think he use the this dataset to analyze the voting history ONLY. This dataset is similar to "countypres_2000-2020.csv"

- Let's look at the election results in the last 10 elections to see what political parties each states voted in the last 10 elections. We want to check a quick survey on which states are the core state for which party. Note that we only use the presidential dataset for this purpose only!

```python
df_elec = pd.read_csv('../data/sample/1976_2016_president.csv')
```

```python
df_elec.head()
```

```python
# extract all state names and abbreviations from the dataset
state_names = collections.defaultdict()
state_abbre = collections.defaultdict()

for i in range(len(df_elec)):
    name = df_elec['state'][i]
    state_names[name] = None
    state_abbre[name] = df_elec['state_po'][i]
```

```python
print(f"There are {len(state_names)} state names and {len(state_abbre)} state abbreviation")
```

```python
def process_row(df):
    # reference: https://github.com/lekhuong07/presidential_election/blob/main/presidential.ipynb
    try:
        first_party = round(df['candidatevotes'][0]/ df['totalvotes'][0]*100, 2)
        second_party = round(df['candidatevotes'][1]/ df['totalvotes'][0]*100, 2)
        third_party = round(df['candidatevotes'][2]/ df['totalvotes'][0]*100, 2)
    except:
        third_party = 0
    finally:
        temp_df = {
            "state_name": [df['state'][0]],  
            "state_abbre": [df['state_po'][0]],
            "year": [df['year'][0]],
            "1st_vote_party": [df['party'][0]],
            "_1st_party": first_party,
            "2nd_vote_party": df['party'][1],
            "_2nd_party": second_party,
            "3rd_vote_party": "Others",
            "_3rd_party": third_party,
            "1st-2nd": first_party - second_party # only the diff 
        }
    retdf = pd.DataFrame(data=temp_df)
    return retdf
```

```python
# reference: https://github.com/lekhuong07/presidential_election/blob/main/presidential.ipynb

processed_df_elec = pd.DataFrame({
    "state_name": ["None"],
    "state_abbre": ["None"],
    "year": [0],
    "1st_vote_party": ["None"],
    "%_1st_party": [0],
    "2nd_vote_party": ["None"],
    "%_2nd_party": [0],
    "3rd_vote_party": ["None"],
    "%_3rd_party": [0],
    "1st-2nd": [0]
})

# go thru selected state and election years only
for sn in state_names.keys():
    for yr in range(1976, 2020, 4):
        # choose the row that have the same name and year
        temp_row = df_elec[(df_elec.state == sn) & (df_elec.year == yr)]
        temp_row = temp_row.reset_index(drop=True)
        processed_temp_row = process_row(temp_row)
        processed_df_elec = pd.concat([processed_df_elec, processed_temp_row], join="inner")
        state_names[sn]

processed_df_elec = processed_df_elec.iloc[1:]
processed_df_elec = processed_df_elec.reset_index(drop=True)
```

```python
processed_df_elec.head()
```

"In five most recent elections, if the number of votes difference for a certain political party is higher than 20% and it happens for 4 out of 5 times. "

```python
# reference: https://github.com/lekhuong07/presidential_election/blob/main/presidential.ipynb
# make copy of dictionary
filled_state_names = state_names.copy()

for sn in state_names.keys():
    democrate_cnt = 0
    republican_cnt = 0
    
    temp_row = processed_df_elec[(processed_df_elec.state_name == sn)]
    temp_row = temp_row.reset_index(drop=True)
    
    # take 5 most recent votes only
    for i in range(5, 10):
        if temp_row['1st-2nd'][i] >= 20:
            if temp_row["1st_vote_party"][i] == 'democrat':
                democrate_cnt += 1
            else:
                republican_cnt += 1
    if democrate_cnt >= 4:
        filled_state_names[sn] = 1
    elif republican_cnt >= 4:
        filled_state_names[sn] = -1
    else:
        filled_state_names[sn] = 0
```

### Process the 2012 Dataset

```python
# set up training dataset using the 2012 training dataset
min_max_scaler = MinMaxScaler(feature_range=(-1,1))
X_train = vote_df_2012[race_list].values
X_train = min_max_scaler.fit_transform(X_train)

# label for model 1 -> predict republican votes
min_max_scaler = MinMaxScaler(feature_range=(-1,1))
y_train_repu = vote_df_2012['votes_gop_2012'].to_numpy().reshape(-1, 1)
y_train_repu = min_max_scaler.fit_transform(y_train_repu)

# label for model 2 -> predict democratic votes
min_max_scaler = MinMaxScaler(feature_range=(-1,1))
y_train_dem = vote_df_2012['votes_dem_2012'].to_numpy().reshape(-1, 1)
y_train_dem = min_max_scaler.fit_transform(y_train_dem)
```

### Process the 2016 Dataset

```python
# set up testing dataset using the 2016 testing dataset
min_max_scaler = MinMaxScaler(feature_range=(-1,1))
X_test = vote_df_2016[race_list].values
X_test = min_max_scaler.fit_transform(X_test)

# label for model 1 -> predict republican votes
min_max_scaler = MinMaxScaler(feature_range=(-1,1))
y_test_repu = vote_df_2016['votes_gop_2016'].to_numpy().reshape(-1, 1)
y_test_repu = min_max_scaler.fit_transform(y_test_repu)

# label for model 2 -> predict democratic votes
min_max_scaler = MinMaxScaler(feature_range=(-1,1))
y_test_dem = vote_df_2016['votes_dem_2016'].to_numpy().reshape(-1, 1)
y_test_dem = min_max_scaler.fit_transform(y_test_dem)
```

### Train and Test the Model for Republican Votes

```python
# create a model for democrate votes prediction
# lr_repu = linear_model.LinearRegression()
lr_repu = sklearn.svm.SVR()

# train model on 2012 data
lr_repu.fit(X_train, y_train_repu)

# test model on 2016 data
y_pred_repu = lr_repu.predict(X_test)

# calculate mse score between prediction and label
mse = metrics.mean_squared_error(y_test_repu, y_pred_repu)

print(mse * 100)
```

```python
# Reference: http://localhost:8888/notebooks/Documents/school/graduate/ece%20_5424/project/ece_5424_project/src/presidential.ipynb
def analyze_state(state_df):
    rep = 0
    dem = 0
    na  = 0  # in case there is a county that is undecided.
    for i in range(len(state_df)):
        if state_df['Won'][i] == "Republican":
            rep += 1
        elif state_df['Won'][i] == 'Democratic':
            dem += 1
        else:
            na += 1
    if dem > rep:
        return 1
    else:
        return -1
```
