### Obtain important features

```python
import pandas as pd
```

```python
df = pd.read_excel("features.xlsx")
df
```

```python
# Compute how many times each feature was selected by total of 6 models
count = df['all'].value_counts()
count
```

```python
# Only keep features that occurred more than 3 times in column "all"
# https://stackoverflow.com/questions/62811654/keep-only-rows-repeated-more-than-four-times-in-a-pandas-dataframe-column
df = df[df['all'].map(df['all'].value_counts()).gt(3)]
df
```

```python
# Drop duplicate values in column "all"
# Rename column "all" to "important_features"
# Show "important_features" only
df = df.drop_duplicates(subset=['all'])
df = df.rename(columns={'all': 'important_features'})
df = df['important_features']
df.tolist()
```

```python
# Create important_featres.xlsx
df.to_excel("important_features.xlsx")
```
